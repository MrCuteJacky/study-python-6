```python
def times2(i):
    return 2 * i
df["comments2"] = df["comments"].apply(times2)
df.head(1)
```

```python
from datetime import datetime
from datetime import date

df["film_date_2_datetime"] = df["film_date"].apply(datetime.fromtimestamp)
df["film_date_2_date"] = df["film_date"].apply(date.fromtimestamp)
df.head(1)
```

```python
df.drop(["film_date_2_datetime"], axis=1)
df.head(1)
```

```python
def add(i, j):
    return i + j
df["comments_duration"] = df.apply(lambda row: add(row['comments'], row['duration']), axis=1)
df["comments_duration_2"] = df.apply(lambda row: row['comments'] + row['duration'], axis=1)
df.head(1)
```

```python
df.head(3)[[True, True, False]]
```

```python
def commentflow(i):
    return i <= 5000 and i >= 3000
f = df["comments"].apply(commentflow)
df[f]
```

```python
def tagflow(i):
    tags = eval(i)
    return "children" in tags

df[ df["tags"].apply(tagflow) ]
```

```python
def tagflow(i, **kwargs):
    tags = eval(i)
    if kwargs["target"] in tags:
        return True
    else:
        return False

s = input("Please input keyword: ")
target_df = df[df["tags"].apply(tagflow, target=s)]
target_df.to_csv("filter.csv", encoding="utf-8", index=False)
target_df
```
